# GET TABLES

```python
import numpy as np
import pandas as pd
import sqlite3
```

### GET

```python
dow_url = 'https://en.wikipedia.org/wiki/Dow_Jones_Industrial_Average'
sp5_url = 'https://en.wikipedia.org/wiki/List_of_S%26P_500_companies'
etf_url = 'https://etfdb.com/compare/volume/'

dow_raw = pd.read_html(dow_url)[1]
sp5_raw = pd.read_html(sp5_url)[0]
etf_raw = pd.read_html(etf_url)[0]
```

### FORMAT

```python
dow_table =  pd.DataFrame({
    'Symbol': dow_raw['Symbol'].str.replace('.', '-')})

sp5_table =  pd.DataFrame({
    'Symbol': sp5_raw['Symbol'].str.replace('.', '-'),
    'Name': sp5_raw['Security'],
    'Sector': sp5_raw['GICS Sector'],
    'Industry': sp5_raw['GICS Sub-Industry']})

dow_table = dow_table.merge(sp5_table)

etf_table =  pd.DataFrame({
    'Symbol': etf_raw['Symbol'],
    'Name': etf_raw['Name']})

display(dow_table)
display(sp5_table)
display(etf_table)
```
